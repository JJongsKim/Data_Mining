```python
import pandas as pd
import matplotlib.pyplot as plt 
from pandas import Series, DataFrame

data = pd.read_csv("nba.csv")

#나이 히스토그램
Age_arr = data['Age'].values
print("1. 나이 히스토그램")
plt.hist(Age_arr)
```

```python
#연봉이 가장 높은 선수
pd.options.display.float_format = '{:.1f}'.format

high = data.sort_values(by='Salary', ascending=0).head(1)
high = high[['Name', 'Team', 'Position', 'Age', 'Salary']]
high_info = high.values
high_info = high_info.ravel()

series_data = pd.Series(high_info, index=['Name', 'Team', 'Position', 'Age', 'Salary'])
print("2. 연봉이 가장 높은 선수의 정보")
series_data
```

```python
#팀별 평균연봉
mean_salary = data['Salary'].groupby(data['Team']).mean()
print("3. 팀별 평균연봉")
mean_salary.plot.bar(x='Team', y='Salary')
```

```python
#선수들의 출신 학교별 인원수
count_student = data['College'].groupby(data['College']).count()
print("4. 출신 학교별 인원수")
print(count_student)
```

```python
#선수들의 평균 연봉과 나이
total_meanAge = data['Age'].mean()
total_meanSalary = data['Salary'].mean()

print("5. 선수들의 평균 연봉과 나이")
print("Avg Age: ", total_meanAge)
print("Avg Salary: ", total_meanSalary)
```

```python
#포지션 별 이름 출력하기
position_name = data['Name'].groupby(data['Position'])

print("6. 선수포지션 별 이름 출력")

for index, data in position_name:
    print(index)
    print(data)
```
